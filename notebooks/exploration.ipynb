```python
import pandas as pd
```

```python
realtor_hotness = pd.read_csv("https://econdata.s3-us-west-2.amazonaws.com/Reports/Hotness/RDC_Inventory_Hotness_Metrics_County_History.csv")
cols = ["county_name", "median_listing_price"]

map_data = realtor_hotness[cols].copy()
map_data["date"] = realtor_hotness.month_date_yyyymm.copy()

map_data.date = pd.to_datetime(map_data.date, format="%Y%m").dt.to_period("M")

def extract_state(col):
    return col.split(", ")[-1].upper()

map_data["state"] = map_data.county_name.apply(extract_state)
latlon = pd.read_csv("../data/processed/us_state_lat_lon_abbreviations.csv")
all_map_data = pd.merge(map_data, latlon, on="state", how="inner")
for_the_map = all_map_data[(all_map_data.date.dt.year == 2023) & (all_map_data.date.dt.month == 6)].groupby("state").mean("median_listing_price")
```
