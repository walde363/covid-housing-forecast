```python
import pandas as pd
import numpy as np
import openpyxl

# base repo info
repo = "https://raw.githubusercontent.com/walde363/covid-housing-forecast/main/"
folder = "data/raw/"

def load_rdc_inventory_files():
    dfs = []

    for i in range(1, 33):
        file_name = f"RDC_Inventory_Core_Metrics_Zip_History{i}.csv"
        url = f"{repo}{folder}{file_name}"
        
        try:
            df = pd.read_csv(url)
            df["source_file"] = file_name  # track origin
            dfs.append(df)
        except Exception as e:
            print(f"Failed: {file_name} | {e}")
    
    return pd.concat(dfs, ignore_index=True)

combined_df = load_rdc_inventory_files()

combined_df.to_csv("../data/processed/RDC_Inventory_Core_Metrics_Zip_History.csv", index=False)
```
